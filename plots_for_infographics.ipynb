```python
import matplotlib.pyplot as plt
import random
import numpy as np

# Generate some random data for five clusters (close to one another)
random.seed(43)
fig, ax =plt.subplots(figsize=[3,4])

# Number of nodes in each cluster
num_nodes = 25

# Generate data for five clusters
cluster1 = [(random.gauss(0.7, 0.1), random.gauss(0.7, 0.1)) for _ in range(num_nodes)]
cluster2 = [(random.gauss(1.2, 0.2), random.gauss(1.2, 0.2)) for _ in range(num_nodes)]
cluster3 = [(random.gauss(1.7, 0.2), random.gauss(1.7, 0.2)) for _ in range(num_nodes)]
cluster4 = [(random.gauss(2.2, 0.2), random.gauss(2.2, 0.2)) for _ in range(num_nodes)]
cluster5 = [(random.gauss(2.7, 0.2), random.gauss(2.7, 0.2)) for _ in range(num_nodes)]

# Combine all data points
data = cluster1 + cluster2 + cluster3 + cluster4 + cluster5

# Generate values from 0 to 1 to represent the gradient of colors
gradient_values = np.linspace(0, 1, len(data))

# Extract x and y coordinates
x, y = zip(*data)

# Create the scatter plot with a gradient of colors using the 'viridis' colormap
plt.scatter(x, y, c=gradient_values, cmap='viridis')

# Remove right and top borders
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')

# Add arrows to left and bottom borders
plt.gca().xaxis.set_ticks_position('bottom')
plt.gca().yaxis.set_ticks_position('left')
plt.gca().spines['bottom'].set_position(('data', 0.4))
plt.gca().spines['left'].set_position(('data', 0.4))

# Calculate midpoint coordinates of the arrow
arrow_midpoint_x =  (2.5 - 0.7) / 2
arrow_midpoint_y = 0.5 + (4 ) / 2

# Add an arrow from (0.5, 0.5) to the maximum x and y values
plt.arrow(0.5, 1, 2, 2.5, color='black', linewidth=1.5,
          head_width=0.1, head_length=0.1, fc='k', ec='k')

# Add text annotation at the midpoint of the arrow, placed directly on the line
plt.text(arrow_midpoint_x, arrow_midpoint_y, 'Difficulty Level', rotation=60, fontsize=12, verticalalignment='center', c='black')
plt.title('Task Embeddings')


# Show the colorbar to indicate the mapping of gradient values to colors
plt.colorbar()

# Show the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import itertools
import numpy as np
import string


fig, ax =plt.subplots(figsize=[3,4])

n_students = 3
n_items = 5
node_size=350
doc_square_size = 25
n_edges = 10
dimmed_transparency = .2
full_transparency = 1
before_training = False

np.random.seed(0)
random.seed(0)

G = nx.Graph()

left_nodes = list(range(n_students))

alphabet= list(string.ascii_lowercase)
n_items = min(n_items,len(alphabet))
right_nodes = alphabet[:n_items]
G.add_nodes_from(left_nodes, bipartite=0)  # 0 indicates the left set
G.add_nodes_from(right_nodes, bipartite=1)  # 1 indicates the right set

pos = nx.bipartite_layout(G, left_nodes, align='vertical')
nx.draw_networkx(G, pos, with_labels=False, node_size=node_size, font_size=10, font_color='black', node_color = "cyan")

# Define edges between nodes
cartesian_product_edges = list(itertools.product(left_nodes, right_nodes))
n_edges = min(n_edges,len(left_nodes)*len(right_nodes))
edges = random.sample(cartesian_product_edges, n_edges)
set_edge_colors = {'red', 'green', 'gray'}
# NB: here setting frequencies of edge types
edge_colors = random.sample(set_edge_colors, n_edges,counts=[n_edges//3+1,n_edges//3+1,n_edges//3+1])
if before_training: 
    edge_transparencies = [dimmed_transparency if color == 'gray' else full_transparency for color in edge_colors]
else:
    edge_transparencies = [full_transparency if color == 'gray' else dimmed_transparency for color in edge_colors]
    edge_colors = [ (i if i!="gray" else np.random.choice(["red","green"])) for i in edge_colors ]





# symbols for left nodes
diversity_symbols = {"♂","♀"}
left_node_shapes = {i:random.sample(diversity_symbols,1)[0]for i in left_nodes} 
left_node_positions = {node: (x, y) for node, (x, y) in pos.items() if node in left_nodes}
for node, shape in left_node_shapes.items():
    plt.text(left_node_positions[node][0], left_node_positions[node][1], shape, fontsize=15, ha="center", va="center")

# Draw right nodes with squares to symbolize documents
right_node_positions = {node: (x, y - 0.05) for node, (x, y) in pos.items() if node in right_nodes}
for node in right_nodes:
    x, y = right_node_positions[node]
    plt.plot(x, y+.05, 's', markersize=doc_square_size , markerfacecolor='white', markeredgecolor='black')


# Identify what nodes represent
students_label_pos = (np.mean([i[0] for i in left_node_positions.values()]), .75)
plt.text(*students_label_pos, "Students", ha="center", va="center", fontsize=12, color="black")
items_label_pos = (np.mean([i[0] for i in right_node_positions.values()]), .75)
plt.text(*items_label_pos, "Tasks", ha="center", va="center", fontsize=12, color="black")

# Draw edges with specified colors, transparencies, and labels
edge_labels = {}
for edge, color, transparency in zip(edges, edge_colors, edge_transparencies):
    label = ''
    if color == 'green':
        label = '✔'
    elif color == 'red':
        label = '✘'
    elif color == 'gray':
        label = '?'
    
    edge_labels[edge] = label
    
    nx.draw_networkx_edges(
        G, 
        pos, 
        edgelist=[edge], 
        edge_color=color, 
        width=5,#[5**i for i in edge_transparencies], 
        alpha=transparency, 
        label=label
        )

# Draw edge labels
nx.draw_networkx_edge_labels(G,pos, edge_labels=edge_labels, label_pos=.2, font_color='black', font_size=10)

plt.axis('off')
plt.show()
```
